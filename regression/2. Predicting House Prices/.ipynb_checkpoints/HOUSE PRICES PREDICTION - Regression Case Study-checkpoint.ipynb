```python
#Handling Outliers
```

```python
numerical_var = data.dtypes.loc[(data.dtypes=='float64')|(data.dtypes=='int64')].index.tolist()
categorical_var = data.dtypes.loc[(data.dtypes=='object')].index.tolist()
```

```python
data_num = data[numerical_var]
data_cat = data[categorical_var]
```

```python
def outlier_capping(x):
    x = x.clip_upper(x.quantile(0.99))
    x = x.clip_lower(x.quantile(0.01))
    return x

data_num=data_num.apply(lambda x: outlier_capping(x))
```

```python
def create_dummies( df, colname ):
    col_dummies = pd.get_dummies(df[colname], prefix=colname, drop_first=True)
    df = pd.concat([df, col_dummies], axis=1)
    df.drop( colname, axis = 1, inplace = True )
    return df

for c_feature in data_cat.columns:
    data_cat[c_feature] = data_cat[c_feature].astype('category')
    data_cat = create_dummies(data_cat , c_feature )
```

```python
data_cat.head()
```

```python
houseData = pd.concat([data_num, data_cat], axis=1)
houseData.head()
```

```python
sns.distplot(np.log(houseData.Price))
```

This cureve is approximatly Normal curve so we can apply liner regression in the data

## Split the data into train & test


```python
train, test = train_test_split(houseData, test_size = 0.3, random_state = 123)
```

```python
print(len(train))
print(len(test))
```

```python
import statsmodels.formula.api as smf
```

```python
my_formula = 'Price~SqFt+Bedrooms+Bathrooms+Offers+Brick_Yes+Neighborhood_West'
```

```python
model = smf.ols(my_formula, data=train).fit()
```

```python
print(model.summary())
```

> ### y = 52.7295 x SqFt + 4222.2404 x Bedrooms + 8134.7120 x Bathrooms - 7380.1914 x Offers + 1.784e+04 x Brick_Yes +2.381e+04 x Neighborhood_West - 2023.0515 

```python
train['pred'] = pd.DataFrame(model.predict(train))
test['pred'] = pd.DataFrame(model.predict(test))
```

```python
# calculate these metrics by hand!
from sklearn import metrics
import numpy as np
import scipy.stats as stats
```

```python
#Train Data
MAPE_train = np.mean(np.abs(train.Price - train.pred)/train.Price )
print(MAPE_train)


RMSE_train = metrics.mean_squared_error(train.Price , train.pred)
print(RMSE_train)

Corr_train = stats.stats.pearsonr(train.Price , train.pred)
print(Corr_train)

print('##################')
#Test Data
MAPE_test = np.mean(np.abs(test.Price - test.pred)/test.Price )
print(MAPE_test)

RMSE_test = metrics.mean_squared_error(test.Price , test.pred)
print(RMSE_test)

Corr_test = stats.stats.pearsonr(test.Price , test.pred)
print(Corr_test)
```

```python
#Conclusion: Model is getting validated
```

```python
#Decile analysis
```

```python
train['Deciles']=pd.qcut(train['pred'],10, labels=False)

train.head()
```

```python
avg_actual = train[['Deciles','Price']].groupby(train.Deciles).mean().sort_index(ascending=False)['Price']
avg_pred = train[['Deciles','pred']].groupby(train.Deciles).mean().sort_index(ascending=False)['pred']
```

```python
Decile_analysis_train = pd.concat([avg_actual, avg_pred], axis=1)
```
